```python
import scipy.stats
import os
import csv
from scipy.special import erf, erfinv
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,1.,101)
prior=x*0+1

prior=prior/prior.sum()


posterior=prior*((1-x)**110)*((x)**140)

posterior=posterior/posterior.sum()

plt.figure()
plt.plot(x,prior)
plt.plot(x,posterior)
```
