```python
import scipy.stats
import os
import csv
from scipy.special import erf, erfinv
import numpy as np
%matplotlib nbagg
import matplotlib.pyplot as plt
```

```python
#Forward problem
r=250
f=0.1


n=np.arange(500)
probn=scipy.stats.poisson.pmf(n,r)
plt.figure()
plt.plot(n,probn)

k=n.copy()
probk=probn*0
for ni in range(len(n)):
    probk+=probn[ni]*scipy.stats.binom.pmf(k,n[ni],f)
    
plt.figure()
plt.plot(k,probk)
```

```python
#Single bayesian

plt.figure()
for r in [100,250,400]:
    f=0.1
    k=15

    n=np.arange(500)
    probn=scipy.stats.poisson.pmf(n,r)
    probn=probn*scipy.stats.binom.pmf(k,n,f)
    probn=probn/probn.sum()

    plt.plot(n,probn)
```

```python
#Double bayesian

r=np.arange(500)
probr=r*0.

plt.figure()
for ri in range(len(r)):
    f=0.1
    k=15

    n=np.arange(500)
    probn=scipy.stats.poisson.pmf(n,r[ri])
    probn=probn*scipy.stats.binom.pmf(k,n,f)
    probn=probn/probn.sum()

    plt.plot(n,probn)
```
